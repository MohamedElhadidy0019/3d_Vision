```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mask_path = 'dl_challenge/005/mask.npy'
mask = np.load(mask_path)
```

```python
mask.shape
```

```python
# or all layers to get background
bacgkround = np.all(mask == 0, axis=0)
```

```python
# add dimesnion to first of bacgkround, the inverse of squeeze
bacgkround = np.expand_dims(bacgkround, axis=0)
plt.imshow(bacgkround[0])

print(bacgkround.shape)
```

```python
# add background to the mask
mask = np.concatenate([bacgkround, mask], axis=0)
```

```python
mask.shape
```

```python
mask_argmax = np.argmax(mask, axis=0)
```

```python
# show mask argmax
plt.imshow(mask_argmax==2, cmap = plt.cm.get_cmap('tab20'))
```
